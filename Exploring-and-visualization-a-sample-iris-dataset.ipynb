```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# for better plots
sns.set(style="whitegrid")
```

##  Loading the Iris Dataset

We load the Iris dataset using seaborn's built-in function.  
The Iris dataset contains measurements of different flower species and is commonly used as 
an introductory dataset in machine learning and data visualization.

```python
iris = sns.load_dataset("iris")
```

##  Scatter Plot: Sepal Length vs Sepal Width

We create a scatter plot colored by species to visualize relationships between
sepal length and width. This helps us see how different species separate in feature space.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x="sepal_length",
    y="sepal_width",
    hue="species",
    data=iris
)
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

##  Histograms for Feature Distribution

We generate histograms for all numerical features to observe their distributions.
This reveals patterns such as symmetry, skewness, and range of values.

```python
iris.hist(figsize=(10, 8))
plt.suptitle("Feature Distributions")
plt.show()
```

##  Box Plot to Identify Outliers

Box plots help us see:
- The median
- Quartiles
- Potential outliers

This is useful for detecting extreme values that could affect analysis.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris.drop("species", axis=1))
plt.title("Box Plot of Iris Features")
plt.show()
```

##  Key Observations & Takeaways

- The scatter plot shows that certain species (like Setosa) can be separated clearly based on feature values.
- Petal length and petal width show stronger separation among species than sepal measurements.
- Histograms show feature value distributions, revealing which features have wide or narrow ranges.
- Box plots indicate the presence of outliers in some features.

Overall, this exploratory analysis provides a solid understanding of the data before applying any
machine learning algorithms.
